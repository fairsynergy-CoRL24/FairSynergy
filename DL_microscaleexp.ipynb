NUM VS FS EXPERIMENTS:
when hardness levels are the same for all robots

```python
# different from the past experiments, the num of robots increasing for fixed resources: 
np.random.seed(42)
num_robots = 10

Tt = 1
Ebar = 1
# print(Tt)
alpha_initt = np.array([[0.025],[0.05],[0.075],[0.1],[0.125],[0.15],[0.175],[0.2],[0.225],[0.25]])
rho_initt = np.array([[0.025],[0.05],[0.075],[0.1],[0.125],[0.15],[0.175],[0.2],[0.225],[0.25]])
Gg_alpha = np.array([[0.5],[0.5],[0.5],[0.5],[0.5],[0.5],[0.5],[0.5],[0.5],[0.5]])
Gg_rho = np.flip(np.array([[0.5],[0.5],[0.5],[0.5],[0.5],[0.5],[0.5],[0.5],[0.5],[0.5]]))
# initialization of the configuration
con = {"E_bar":Ebar,
        "T":Tt,  
        "alpha_init_vec":alpha_initt,
        "rho_init_vec":rho_initt,
        "gamma2":Gg_alpha,
        "gamma1":Gg_rho,
        "optforrho":True,
        "optforalpha":True
}

# calculation: check the fs allocations
rho_alloc, alpha_alloc, _ = fairSynergy2D(con, num_robots)
rho_initt = con["rho_init_vec"]
alpha_initt = con["alpha_init_vec"]

# plot the allocated resources
ax, fig = plot_bar_DL(rho_initt, alpha_initt, rho_alloc, alpha_alloc, "Amount of Edge Device Resources")
# set the limits
ax.set_ylim([0,0.5])

# calculation: check the allocation performance with fs
nominalAccuracy = calcAcc(rho_initt, alpha_initt, con["gamma1"], con["gamma2"])
fsAccuracyRho = calcAcc( rho_initt+rho_alloc,alpha_initt, con["gamma1"], con["gamma2"]) # TODO check whether the shape is as expected 
fsAccuracy = calcAcc( rho_initt+rho_alloc,alpha_initt+alpha_alloc, con["gamma1"], con["gamma2"]) # TODO check whether the shape is as expected
# plot the resultin accuracy
ax, fig = plot_util_DL(nominalAccuracy, fsAccuracyRho-nominalAccuracy, fsAccuracy-fsAccuracyRho, "Model Accuracy") 

# gamma indicators 


ax.text(4.5, 0.47, "$\gamma_1$=0.5", ha='center', color='gray')
ax.text(4.5, 0.44, "$\gamma_2$=0.5", ha='center', color='gray')


# nominal final accuracy indicators
ax.axhline(y=np.mean(nominalAccuracy), color='blue', linestyle='--')
ax.text(1.5, np.mean(nominalAccuracy)+0.005, "Nominal Average= "+str(np.mean(nominalAccuracy).round(2)), ha='center', color='blue')
ax.axhline(y=np.mean(fsAccuracyRho), color='purple', linestyle='--')
ax.text(1.5, np.mean(fsAccuracyRho)+0.005, "$\\rho\\uparrow$ Average= "+str(np.mean(fsAccuracyRho).round(2)), ha='center', color='purple')
ax.axhline(y=np.mean(fsAccuracy), color='green', linestyle='--')
ax.text(1.5, np.mean(fsAccuracy)+0.005, "Final Average= "+str(np.mean(fsAccuracy).round(2)), ha='center', color='green')

# set the limits
ax.set_ylim([0,0.5])

# calculation: check the num allocations
rho_alloc, alpha_alloc = NUM2D(con, num_robots=num_robots)

# plot the allocated resources
ax, fig = plot_bar_DL(rho_initt, alpha_initt, rho_alloc, alpha_alloc, "Amount of Edge Device Resources")
# set the limits
ax.set_ylim([0,0.5])
# calculation: check the allocation performance with num
nominalAccuracy = calcAcc(rho_initt, alpha_initt, con["gamma1"], con["gamma2"])
fsAccuracyRho = calcAcc( rho_initt+rho_alloc,alpha_initt, con["gamma1"], con["gamma2"]) # TODO check whether the shape is as expected 
fsAccuracy = calcAcc( rho_initt+rho_alloc,alpha_initt+alpha_alloc, con["gamma1"], con["gamma2"]) # TODO check whether the shape is as expected

# plot num allocations
ax, fig = plot_util_DL(nominalAccuracy, fsAccuracyRho-nominalAccuracy, fsAccuracy-fsAccuracyRho, "Model Accuracy") 


# gamma indicators 

ax.text(4.5, 0.47, "$\gamma_1$=0.5", ha='center', color='gray')
ax.text(4.5, 0.44, "$\gamma_2$=0.5", ha='center', color='gray')


# nominal final accuracy indicators
ax.axhline(y=np.mean(nominalAccuracy), color='blue', linestyle='--')
ax.text(1.5, np.mean(nominalAccuracy)+0.005, "Nominal Average= "+str(np.mean(nominalAccuracy).round(2)), ha='center', color='blue')
ax.axhline(y=np.mean(fsAccuracyRho), color='purple', linestyle='--')
ax.text(1.5, np.mean(fsAccuracyRho)+0.005, "$\\rho\\uparrow$ Average= "+str(np.mean(fsAccuracyRho).round(2)), ha='center', color='purple')
ax.axhline(y=np.mean(fsAccuracy), color='red', linestyle='--')
ax.text(1.5, np.mean(fsAccuracy)+0.005, "Final Average= "+str(np.mean(fsAccuracy).round(2)), ha='center', color='red')

ax.set_ylim([0,0.5])
```
