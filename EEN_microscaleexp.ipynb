What is corresponding utility plot for same gamma values

```python
# check the fs allocations
irobots = 10
np.random.seed(42)
Tt = 0.46
Ee = np.array([[0.1],[0.3], [0.5], [0.7], [0.1],[0.3], [0.5], [0.1],[0.3], [0.5]])/10
Gg = np.ones((irobots,1))*0.3

# calculation: check the fs allocations
con = {"T":Tt, "E_init":Ee, "gamma_vec":Gg}
fs_alloc, _ = fairSynergy1D(con, irobots)

# plot the allocated resources
ax, fig = plot_barr(np.round(Ee*100), np.round(fs_alloc*100))
# set the limits
ax.set_ylim([0,12])

# calculation: check the allocation performance with fs
fs = calcAcc1D(fs_alloc+Ee, Gg)
nominal = calcAcc1D(Ee, Gg)
ax, fig = plot_util(nominal, fs-nominal)

# gamma indicators 


ax.text(4.5, 0.55, "$\gamma$=0.3", ha='center', color='gray')


# nominal final accuracies
ax.axhline(y=np.mean(nominal), color='gray', linestyle='--')
ax.text(1.5, np.mean(nominal)+0.005, "Nominal Average= "+str(np.mean(nominal).round(2)), ha='center', color='gray')
ax.axhline(y=np.mean(fs), color='green', linestyle='--')
ax.text(1.5, np.mean(fs)+0.005, "Final Average= "+str(np.mean(fs).round(2)), ha='center', color='green')

# set the limits
ax.set_ylim([0,0.6])

# change the allocation type to num which assumes the gamma are the same
con = {"T":Tt, "E_init":Ee, "gamma_vec":np.ones((irobots,1))*0.6}
fs_alloc, _ = fairSynergy1D(con, irobots)
fs1 = calcAcc1D(Ee+fs_alloc, np.ones((irobots,1))*0.6)

# plot the allocations in bar plot
ax, fig = plot_barr(np.round(Ee*100), np.round(fs_alloc*100))
# set the limits
ax.set_ylim([0,12])
# Check the allocation performance with num
con = {"T":Tt, "E_init":Ee, "gamma_vec":Gg}
fs = calcAcc1D(Ee+fs_alloc, Gg)
nominal = calcAcc1D(Ee, Gg)
ax, fig = plot_util(nominal, fs-nominal)

# gamma indicators 


ax.text(4.5, 0.55, "$\gamma$=0.3", ha='center', color='gray')

# nominal final accuracies
# bbox_props = dict(boxstyle="round,pad=0.3", facecolor="black", edgecolor="gray", alpha=0.05)
ax.axhline(y=np.mean(nominal), color='gray', linestyle='--')
ax.text(1.5, np.mean(nominal)+0.005, "Nominal Average= "+str(np.mean(nominal).round(2)), ha='center', color='gray')#,bbox=bbox_props)
ax.axhline(y=np.mean(fs), color='red', linestyle='--')
ax.text(1.5, np.mean(fs)+0.005, "Final Average= "+str(np.mean(fs).round(2)), ha='center', color='red')

# set the limits
ax.set_ylim([0,0.6])
```
