```python
results_fs, results_uni, results_rand = boxplot_experiments()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data for three x-variables
data = {
    'Random Allocation': results_rand,
    'Uniform Allocation': results_uni,
    'FairSynergy': results_fs
}

# Create a DataFrame from the data
import pandas as pd
df = pd.DataFrame(data)

# Plotting
sns.set()  # Set the style of the plot
plt.figure(figsize=(8, 6))  # Set the size of the plot

# Create the boxplot with three x-variables
sns.boxplot(data=df, whis=1.5)

# Set labels and title
plt.xlabel('Allocation Type', fontsize=14)
plt.ylabel('Aggregate Accuracy', fontsize=14)
plt.title('Cloud Assisted Distributed Learnig', fontsize=16)
plt.ylim([0.45, 0.85])
# Show the plot
plt.show()
```
